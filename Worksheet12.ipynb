# Worksheet: Shortest-path problem

In graph theory, the shortest path problem is the problem of finding a path between two vertices (or nodes) in a graph such that the sum of the weights of its constituent edges is minimized. It is more interesting to consider the weighted directed graph.

![](https://upload.wikimedia.org/wikipedia/commons/3/3b/Shortest_path_with_direct_weights.svg)

Shortest path (A, C, E, D, F) between vertices A and F in the weighted directed graph.


The shortest-path problem has many applications such as robotics and transportation.

## Create graph: Create the graph in above image and then visualize it.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Step 1: Create the weighted directed graph
G = nx.DiGraph()  # Directed graph

# Add nodes and edges with weights
edges_with_weights = [
    ('A', 'C', 2),
    ('C', 'E', 3),
    ('E', 'D', 4),
    ('D', 'F', 11),
    ('A', 'B', 4),
    ('B', 'C', 5),
    ('B', 'D', 10),
]
G.add_weighted_edges_from(edges_with_weights)
```

## Shortest-path

Read the documentation in nextworkx: https://networkx.org/documentation/stable/reference/algorithms/shortest_paths.html. Then, apply Dijkstra's algorithm to find the shortest-path from A to F. 

Dijkstra's algorithm is an algorithm for finding the shortest paths between nodes in a weighted graph. Please search this algorithm and answer why it is the best algorithm.

```python
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='blue', node_size=2000, font_size=15)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Weighted Directed Graph")
plt.show()

# Step 3: Find the shortest path using Dijkstra's algorithm
shortest_path = nx.shortest_path(G, source='A', target='F', weight='weight', method='dijkstra')
shortest_path_length = nx.shortest_path_length(G, source='A', target='F', weight='weight')

print("Shortest path from A to F:", shortest_path)
print("Shortest path length from A to F:", shortest_path_length)
```
